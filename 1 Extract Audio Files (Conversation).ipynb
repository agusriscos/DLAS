```python
import os
import time
import subprocess
```

```python
destination_folder = './final_folder'
```

```python
if not os.path.exists(destination_folder):
    os.makedirs('final_folder')
```

```python
for root, dirs, files in os.walk('./wav'):
    for name in files:
        if name == '.DS_Store':
            continue
        os.rename(os.path.join(root, name),os.path.join(destination_folder, str(round(time.time()*1000)) + '_conversation.wav'))
```

```python
cmd_string = 'ffmpeg -i {tr} -f segment -segment_time 2 -c copy {out}'

destination_folder = './conv_parts'
if not os.path.exists(destination_folder):
    os.makedirs('conv_parts')
    
for root, dirs, files in os.walk('./Data'): # Modificar aqui la ruta
    for f in files:
        if(f == '.DS_Store'):
            continue
        output = './conv_parts/' + f.split('_')[0] + '%1d_conversation.wav' # Modificar aqui la clase y la ruta
        command = cmd_string.format(tr=os.path.join(root, f), out=output)
        print(command)
        subprocess.call(command, shell=True)
```
